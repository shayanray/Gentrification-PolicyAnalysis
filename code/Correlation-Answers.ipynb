```python
import pandas as pd
import numpy as np
```

```python
cols_to_drop = ['TRACTCE10', 'Pop_Src', 'Race_Src', 'Elders_Rpt','Elders_Src', 'SpksEng_Rp',
                'SpksEng_Sr', 'Emp_Src' , 'Income_Rpt', 'Income_Src', 'county', 'tract_FIPS', 'tract2'  ]
bike_gentrification_correlation_df = bike_gentrification_correlation_df.drop(cols_to_drop, axis=1)
```

```python
# set it to the correct datatype 
bike_gentrification_correlation_df['GEOID10'] = bike_gentrification_correlation_df['GEOID10'].astype('category')
bike_gentrification_correlation_df['COUNTYFP10'] = bike_gentrification_correlation_df['COUNTYFP10'].astype('category')
bike_gentrification_correlation_df['Cnty_Name'] = bike_gentrification_correlation_df['Cnty_Name'].astype('category')
bike_gentrification_correlation_df['City_Name'] = bike_gentrification_correlation_df['City_Name'].astype('category')
bike_gentrification_correlation_df['ZIP'] = bike_gentrification_correlation_df['ZIP'].astype('category')


bike_gentrification_correlation_df['race00'] = bike_gentrification_correlation_df['race00'].astype('category')
bike_gentrification_correlation_df['race90'] = bike_gentrification_correlation_df['race90'].astype('category')
bike_gentrification_correlation_df['race15'] = bike_gentrification_correlation_df['race15'].astype('category')

bike_gentrification_correlation_df['disadv'] = bike_gentrification_correlation_df['disadv'].astype('category')
bike_gentrification_correlation_df['gent'] = bike_gentrification_correlation_df['gent'].astype('category')
bike_gentrification_correlation_df['type'] = bike_gentrification_correlation_df['type'].astype('category')


# convert to int

bike_gentrification_correlation_df['disadvantaged_improved'] = bike_gentrification_correlation_df['disadvantaged_improved'].replace({True: 1, False: 0})
bike_gentrification_correlation_df['disadvantaged_worsened'] = bike_gentrification_correlation_df['disadvantaged_worsened'].replace({True: 1, False: 0})
bike_gentrification_correlation_df['disadvantaged_sustained_high'] = bike_gentrification_correlation_df['disadvantaged_sustained_high'].replace({True: 1, False: 0})
bike_gentrification_correlation_df['disadvantaged_sustained_low'] = bike_gentrification_correlation_df['disadvantaged_sustained_low'].replace({True: 1, False: 0})

bike_gentrification_correlation_df['gentrification_improved'] = bike_gentrification_correlation_df['gentrification_improved'].replace({True: 1, False: 0})
bike_gentrification_correlation_df['gentrification_worsened'] = bike_gentrification_correlation_df['gentrification_worsened'].replace({True: 1, False: 0})
bike_gentrification_correlation_df['gentrification_sustained_high'] = bike_gentrification_correlation_df['gentrification_sustained_high'].replace({True: 1, False: 0})
bike_gentrification_correlation_df['gentrification_sustained_low'] = bike_gentrification_correlation_df['gentrification_sustained_low'].replace({True: 1, False: 0})
```
