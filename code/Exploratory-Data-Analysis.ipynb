```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("/content/Bike_gentrification_correlation_v1.csv")
df["has_bikeroute"]=df["All_lengths"]>0
```

```
df.columns.values
```

```
df = pd.get_dummies(df, columns=['neighbourhood_type'])
df = df.rename(columns={"neighbourhood_type_Established suburb": 'Established suburb', "neighbourhood_type_Mixed-use": 'Mixed-use',
                        "neighbourhood_type_New development": 'New development',"neighbourhood_type_Old urban": 'Old urban',
                        "neighbourhood_type_Patchwork": 'Patchwork', "neighbourhood_type_Rural": 'Rural',
                        "neighbourhood_type_Urban residential": 'Urban residential'})
```

```
cols= ['numinstall_bikelanes_Class0','numinstall_bikelanes_Class1',
       'numinstall_bikelanes_Class2', 'numinstall_bikelanes_Class3',
       'numinstall_bikelanes_Class4',
       "All_lengths","Gentrified_1990_2000","Gentrified_2000_2015",
       "Established suburb","Mixed-use",	"New development",
       "Old urban","Patchwork",	"Rural",	"Urban residential"]
corr = df[cols].corr()
sns.heatmap(corr, cmap="coolwarm", center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, vmin=-1, vmax=1)
```
